## First things first

```python
import youtube_dl # Command-line program to download videos from YouTube.com and other video sites

import pandas as pd # for data manipulation
import numpy as np # for numercial operations
```

The library youtube_dl is extremely useful to extract the complete video info or even download them, this is the easiest and quicly way to collect the data from youtube. Let's create the loop to iterate over our key words (storage in the queries list) ordered by upload date, and then create the object ydl to extract the info of the videos suggested for each key word:

```python
queries = ['machine+learning', 'data+science', 'kaggle']
```

```python
ydl = youtube_dl.YoutubeDL({'ignoreerrors': True}) # ignoreerrors avoid problems with unavailables videos

videos_info = []
for query in queries:
    each_video_info = ydl.extract_info("ytsearchdate2000:{}'".format(query), download = False) # Extract the info of 30 videos sorted by date, without do download of them 
    for entry in each_video_info['entries']: # the key 'entries' contains the relevant info concerning each video
        try:
            entry['query'] = query # adding the key:value pair for the query the video belongs
        except:
            pass
    videos_info += each_video_info['entries'] # list where each element is a dictionary of infos about one video
```

```python
videos_info = [element for element in videos_info if element is not None] # picking up just the successful data
```

```python
df = pd.DataFrame(videos_info) #transforming into a dataframe
df.to_feather(r'Raw_Data/raw_data.feather')
```
