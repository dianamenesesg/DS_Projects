```python
import os, sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")

import pandas as pd

main_path = os.getcwd()
```

```python
colunas = ['Data_Sessão',
'Símbolo_do_Instrumento',
'Sentido_Of_Compra',
'Sequência',
'GenerationID_Of_Compra',
'Cód_do_Evento_da_Of_Compra',
'Hora_Prioridade',
'Ind_de_Prioridade_Of_Compra',
'Preço_Of_Compra',
'Qtd_Total_Of_Compra',
'Qtd_Negociada_Of_Compra',
'Data_Oferta_Compra',
'Data_de_Entrada_Of_Compra',
'Estado_Of_Compra',
'Condição_Oferta',
'Corretora']
```

```python
list_instrumentos = ['SUZB3                                             ',
'MGLU3                                             ',
'FIBR3                                             ',
'BTOW3                                             ',
'VALE3                                             ',
'BRKM5                                             ',
'EMBR3                                             ',
'PETR4                                             ',
'CPFE3                                             ',
'KLBN4                                             ',
'BRFS3                                             ',
'KROT3                                             ',
'ELET6                                             ',
'QUAL3                                             ',
'UGPA3                                             ',
'ECOR3                                             ',
'ELET3                                             ',
'CCRO3                                             ',
'SBSP3                                             ',
'GOLL4                                             ']
```

```python
#period = [(10,30), (7, 6), (7, 13)]
#period = [(10,29), (10,8), (11,14),(11,16),(10,22),(10,5),(10,16),(8,7),(9,6),(7,31),(8,22),(9,21),(9,5),(8,28),(11,22)]
#period = [(7,11),(7,13),(7,20),(7,25),(8,10),(8,14),(8,16),(8,21),(8,31)]

period = [(9,11),(9,14),(9,24),(9,27),(10,26)]
for per in period:
    mth, day = per
    load_path = "Bases\\Drty\\NEW\\OFER_CPA_2018{}{}.txt".format(str(mth).zfill(2), str(day).zfill(2))
    load_path = os.path.join(main_path, load_path)
    save_path = "Bases\\CPA\\OFER_CPA_2018{}{}.txt".format(str(mth).zfill(2), str(day).zfill(2))
    save_path = os.path.join(main_path, save_path)
    print(load_path)
    df_cpa = pd.DataFrame()
    for chunk in pd.read_csv(load_path, sep=";",low_memory=False,header=1, chunksize=1000000):
        df_cpa = pd.concat([df_cpa, chunk])
        #print('chunk')
    df_cpa.columns = colunas
    df_cpa = df_cpa[df_cpa["Símbolo_do_Instrumento"].isin(list_instrumentos)] # elimina assets no estudiados
    df_cpa = df_cpa[df_cpa['Cód_do_Evento_da_Of_Compra']==4]
    df_cpa.drop(['Data_Sessão',

    'Sentido_Of_Compra',
    'Sequência',
    'GenerationID_Of_Compra',
    'Cód_do_Evento_da_Of_Compra',

    'Ind_de_Prioridade_Of_Compra',

    'Qtd_Total_Of_Compra',
    'Qtd_Negociada_Of_Compra',
    'Data_Oferta_Compra',
    'Data_de_Entrada_Of_Compra',
    'Estado_Of_Compra',
    'Condição_Oferta',
    'Corretora'],inplace = True, axis = 1)
    
    df_cpa.to_csv(save_path, index=False)
    print(load_path, save_path)
```
